# Exploratory Data Analysis (EDA)

```python
# What is EDA?

# EDA stands for Exploratory Data Analysis.

# In simple words: EDA is the step where we look closely at our data before doing anything else with it.
# We check what the data looks like, find missing values, spot weird/unusual entries, and understand the patterns hiding inside the dataset.

# Think of it like this: before cooking a meal, you first check what ingredients you have, whether anything is missing, and whether anything looks off. EDA is that "checking" step for data.

# In short: EDA = Look at the data -> Understand it -> Clean it -> Get it ready for the next step (modeling).
```

4. Handling Missing Values in EDA

```python
## How do we actually deal with missing values? It depends on the column type.

## Categorical / Text Data
# Option 1: Fill with the Mode (the most common value) -> usually the safest choice
# Option 2: Fill with a placeholder like "Unknown"
# Option 3: Forward Fill (carry the previous value down)
# Option 4: Backward Fill (carry the next value up)

## Numeric Data (int, float)
# Option 1: Fill with the Mean -> works well when data is roughly symmetric
# Option 2: Fill with the Median -> better when data is skewed / has extreme values
# Option 3: Fill with a fixed constant
# Option 4: Interpolation (estimate based on neighboring values)

## Boolean Data
# Option 1: Fill with the Mode
# Option 2: Fill with False
# Option 3: Fill with True

### NaN simply means "Not a Number" -> it's how pandas marks a missing value

## Why the mean for numeric columns? Because it keeps the filled value within a sensible range
## instead of introducing something way outside the normal spread of the data.
```

**Quick Refresher: Mean, Median, Modet**

```python
## Mean, Median, Mode - explained simply

# Mean: the average value. Add every number together, then divide by how many numbers there are.

# Median: the middle value once the numbers are sorted smallest to largest.
# Example: 4, 6, 9, 12 -> the two middle numbers are 6 and 9 -> median = (6+9)/2 = 7.5

# Mode: the value that shows up most often in the data.
# Example: 4, 6, 8, 8, 8, 10, 12, 10 -> mode = 8 (appears three times, more than any other value)
# If there's a tie, there can be more than one mode - mode()[0] simply grabs the first one.
```

**What is Data Wrangling?**

```python
# Data Wrangling (also called data munging) is the process of taking raw, messy data and reshaping it into a clean, structured, ready-to-use format.

## EDA and Data Wrangling usually happen side by side, not one strictly after the other.
```

 EDA vs. Wrangling - What's the Difference?

```python
## Exploring the data (EDA) means figuring out what's WRONG or missing in it.

## Wrangling means actually FIXING those problems.

# EDA            = "What's going on with my data?"
# Data Wrangling  = "How do I fix it?"



## EDA (Explore)              Data Wrangling (Fix)
# Spot the problem              Solve the problem
# Count missing values          Fill in / drop missing values
# Check for duplicates          Remove duplicates
# Detect outliers               Treat / remove outliers
# Understand the data           Prepare the data for modeling
```

**Outlier Detection & Handling (Quick List)**

```python
## Ways to SPOT Outliers

# 1. Box Plot
# 2. Scatter Plot
# 3. Histogram
# 4. IQR (Interquartile Range) Method -> most commonly used
# 5. Z-Score Method
# 6. Modified Z-Score
# 7. Percentile Method
# 8. Isolation Forest (Machine Learning)
# 9. Local Outlier Factor - LOF (Machine Learning)
# 10. DBSCAN (Clustering-based)


## Ways to HANDLE Outliers

# 1. Remove them -> simply delete the rows containing outliers
# 2. Capping / Winsorization -> cap extreme values at an upper/lower limit instead of deleting
# 3. Transform the data
#    ->  Log Transformation
#    ->  Square Root Transformation
#    ->  Box-Cox Transformation
#    ->  Yeo-Johnson Transformation
# 4. Replace with Mean/Median -> swap extreme values instead of removing them outright
# 5. Use Robust Models -> pick algorithms that are naturally less sensitive to outliers(e.g., tree-based models like Decision Trees / Random Forest)
```
