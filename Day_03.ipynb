# Data Wrangling

Data wrangling is the process of cleaning, transforming, and organizing data so that it can be used for analysis and visualization. It is an important step in the data analysis process because raw data is often incomplete, inconsistent, and in an unstructured format, which can make it difficult to work with. Data wrangling helps to make the data more consistent, accurate, and useful for analysis and decision making.

 * Data Wrangling helps to make the data more consistent, accurate and useful for analysis and decision making.

 ### The steps for data wrangling typically include:

1.Gathering data: Collecting data from various sources such as databases, APIs, and flat files.

Assessing data: Examining the data to identify any issues such as missing values, outliers, and duplicate records.

Cleaning data: Removing or correcting any errors, inconsistencies, and irrelevant information from the data. This step often includes tasks such as filling in missing values, dropping duplicate records, and correcting data types.

Transforming data: Modifying the data in order to make it more useful for analysis. This step often includes tasks such as creating new variables, aggregating data, and normalizing data.

Organizing data: Storing the data in a structured format that is easy to access and analyze. This step often includes tasks such as creating a data dictionary, creating a data model, and storing the data in a database.

Analyzing Data: Once the data is cleaned, transformed and organized then we can use this data for different

It's also worth noting that data wrangling is an iterative process and may require multiple rounds

  1. Visualization

```python
import seaborn as sns
plt.figure(figsize=(6,4))
sns.boxplot(data=df, y='age', x='sex')
```

  2. IQR method

```python
Q1= df['age'].quantile(0.25)
Q3= df['age'].quantile(0.75)
IQR = Q3 - Q1
IQR
```

```python
Q1= df['age'].quantile(0.25)
Q3= df['age'].quantile(0.75)
IQR = Q3 - Q1
IQR
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df = df[(df['age'] > lower_bound) & (df['age'] < upper_bound )]
```

```python
df.shape
```

```python
Q1= df['fare'].quantile(0.25)
Q3= df['fare'].quantile(0.75)
IQR = Q3 - Q1
IQR
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df = df[(df['fare'] > lower_bound) & (df['fare'] < upper_bound )]
```
